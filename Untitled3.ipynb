```python
df_fresh = pd.read_csv('student_dropout_dataset_v3.csv')
print("Missing values in raw data:")
print(df_fresh.isnull().sum())
```

```python
# --- Task 1 & 2: Handling Missing Values (Median Imputation)
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
print("--- Task 1: Data Quality Issues ---")
missing_before = df.isnull().sum().sum()
print(f"Total missing values found: {missing_before}")
print(f"Task 2 Complete: Missing values after imputation: {df.isnull().sum().sum()}")

df['Parental_Education'] = df['Parental_Education'].fillna(df['Parental_Education'].mode()[0])

print(df[['Family_Income', 'Study_Hours_per_Day', 'Stress_Index', 'Parental_Education']].isnull().sum())
```

```python
# --- Task 3: Outliers Detection & Handling Results ---
df_raw = pd.read_csv('student_dropout_dataset_v3.csv')

print("--- Task 3: ---")

for col in ['Family_Income', 'Study_Hours_per_Day', 'Attendance_Rate']:

    Q1 = df_raw[col].quantile(0.25)
    Q3 = df_raw[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers_before = ((df_raw[col] < (Q1 - 1.5 * IQR)) | (df_raw[col] > (Q3 + 1.5 * IQR))).sum()


    outliers_after = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()

    print(f"Feature: {col}")
    print(f" - Before Cleaning: {outliers_before} outliers")
    print(f" - After Cleaning: {outliers_after} outliers")
    print("-" * 30)
```

```python
# --- Task 4: Normalization (Min-Max and Z-score) ---
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_min_max = df.copy()
df_min_max[numeric_cols] = min_max_scaler.fit_transform(df[numeric_cols])

# Z-score Scaling (Standardization)
std_scaler = StandardScaler()
df_z_score = df.copy()
df_z_score[numeric_cols] = std_scaler.fit_transform(df[numeric_cols])

print("\n--- Task 4: Normalization Samples ---")
print("Min-Max (First 2 rows):\n", df_min_max.head(2))
print("\nZ-score (First 2 rows):\n", df_z_score.head(2))
```

```python
# --- Task 5: PCA (Apply if correlation exists) ---

print("\n--- Task 5: PCA Analysis ---")
correlation_matrix = df[['GPA', 'Semester_GPA', 'CGPA']].corr()
print("Correlation Matrix:\n", correlation_matrix)

pca = PCA(n_components=1)
df['GPA_Feature_Reduced'] = pca.fit_transform(df_z_score[['GPA', 'Semester_GPA', 'CGPA']])
print("\nPCA Component (First 5 rows):\n", df[['GPA_Feature_Reduced']].head())
```

```python
# ---Task 1: Identify Data Quality Issues
# ---After inspecting the raw dataset, the following data quality issues were identified:

# ---Missing Values: A significant number of missing entries were found in key features:

# ---Family_Income: 500 missing values.

# ---Study_Hours_per_Day: 500 missing values.

# ---Stress_Index: 500 missing values.

# ---Parental_Education: 511 missing values.

# ---Redundancy: High correlation between GPA-related columns.

# ---Scale Inconsistency: Numerical values like Family_Income (thousands) and GPA (0-4) are on completely different scales.
```

```python
# ---Task 2: Missing Value Strategy
# ---Strategy Applied: Median Imputation.
Strategy: 1.  For Numerical columns (Family_Income, Study_Hours, Stress_Index), we applied Median Imputation.
2.  For the Categorical column (Parental_Education), we applied Mode Imputation (replacing with the most frequent value).

Explanation (Why):

Median: Chosen for numerical data because it is not affected by outliers. For example, if a few students have extremely high family incomes, the "Mean" would be skewed, but the "Median" remains a fair representation of the average student.

Mode: Since Parental_Education is a text-based (categorical) category, we cannot calculate a mathematical average. Therefore, filling gaps with the most common education level (Mode) is the most logical way to maintain the dataset's distribution.
```

```python
Detection: Initially, we identified outliers in several columns, such as Family_Income (604 outliers) and Study_Hours_per_Day (86 outliers), using the Interquartile Range (IQR) method.Handling: We applied Capping. This method was chosen to retain the size of the dataset while mitigating the influence of extreme values.Verification: After running the cleaning function,
a re-check was performed, confirming that 0 outliers remain in the dataset, as all values now fall within the calculated safe range $[Q1 - 1.5 \times IQR, Q3 + 1.5 \times IQR]$.
```

```python
Task 4: Numerical Feature Normalization Techniques: Both Min-Max Scaling and Z-score Standardization were applied.Results:
 * Min-Max: Rescaled all features to a fixed range of $[0, 1]$, making them suitable for distance-based algorithms like K-Nearest Neighbors.Z-score:
 Transformed the data to have a mean of $0$ and a standard deviation of $1$, which is the standard requirement for most gradient-based machine learning models.
```

```python
Task 5: Principal Component Analysis (PCA)
Correlation Evidence: A very high positive correlation (approx. 0.96 to 0.99) was detected between academic performance metrics: GPA, Semester_GPA, and CGPA.

Implementation: Due to this high redundancy, PCA was applied to reduce these three dimensions into a single "Principal Component" (GPA_Feature_Reduced).

Explanation (Why): This step was crucial to eliminate Multicollinearity. By compressing these correlated features into one, we reduce the complexity of the model and prevent "Overfitting" while still retaining over 95% of the information (variance) from the original columns.
```
